```python
#********************************************************************
#                    DATE STRTED:                                   *
#                   01 November 2022 [GMT-2]                        *
#                                                                   *
#     Author: Ipfani Mutavhatsindi                                  *
#     LinkedIn: https://www.linkedin.com/in/ipfani-mutavhatsindi    *       
#     Website: https://ipfani.github.io/                            *    
#                                                                   *
#********************************************************************

# The main aim of this project is to get the PSL log data table, clean it and then use ML to predict the outcome of the next match

from bs4 import BeautifulSoup
import requests
import pandas as pd 
import urllib.parse 
import pprint

from datetime import datetime

# Returns the current local date
current_dateTime = date.today()


pp = pprint.PrettyPrinter(indent=10)

#Website PSL WEBSITE
website =requests.get('https://www.psl.co.za/matchcentre?type=log')


#My variables for storing raw data
team_names = []
team_logos = []
match_played = []
win = []
Draw = []
Lost = []
Points = []

#Lets create DataFrame with these headings
df_psl = pd.DataFrame(columns=['Name', 'Match Played', 'Win', 'Draw', 'Lost', 'Points'])

soup = BeautifulSoup(website.content, 'html.parser')

table_html = soup.find('tbody', id="LogViewContent")
## pp.pprint(table_html)

# Every team is divided in a row (tr), This gives me team data as in rows
##team_row_data = table_html.find_all('tr')
## pp.pprint(team_row_data)

#''' Output of a row from the PSL website
    # <tr>
    # <td class="logs-team">
    # <div class="team-meta">
    # <a href="/club/Mamelodi_Sundowns?league=betway-premiership">
    # <div class="team-meta__info">
    # <h5 class="team-meta__name">2   </h5>
    # </div>
    # <figure class="team-meta__logo">
    # <img alt="Mamelodi Sundowns" onerror="this.src='https://www.psl.co.za/newpsl/images/clublogos/noclublogo.png';" src="https://www.psl.co.za/newpsl/images/clublogos/Mamelodi Sundowns.png"/>
    # </figure>
    # <div class="team-meta__info">
    # <h6 class="team-meta__name">Mamelodi Sundowns</h6>
    # </div>
    # </a>
    # </div>
    # </td>
    # <td class="logs-played">6</td>
    # <td class="logs-win">5</td>
    # <td class="logs-draw">0</td>
    # <td class="logs-lost">1</td>
    # <td class="logs-goals-for">13</td>
    # <td class="logs-goals-against">4</td>
    # <td class="logs-goal-diff">9</td>
    # <td class="logs-points">15</td>
    # </tr>,
#'''

team_names = soup.find_all(class_= 'team-meta__name')
match_played = soup.find_all('td', class_='logs-played')
Win = soup.find_all('td', class_='logs-win')
Draw = soup.find_all('td', class_='logs-draw')
Lost = soup.find_all('td', class_='logs-lost')
Points = soup.find_all('td', class_='logs-points')


#OUTPUT
#******************************************************************************************************************************************** 

for row in range (16): # There are 16 teams in PSL as of November 2024
    df_psl = pd.concat([df_psl, pd.DataFrame.from_records([{'Name':team_names[row].string,
                                                            'Match Played' :match_played[row].string,
                                                            'Win':Win[row].text,
                                                            'Draw':Draw[row].string,
                                                            'Lost':Lost[row].string,
                                                            'Points':Points[row].string}])], ignore_index=True)



#******************************************************************************************************************************************** 

print("***********************************************************************")
print("Current Table Log as of ", current_dateTime)
print("***********************************************************************")
print(" ")

df_psl.info()
print(" ")

pp.pprint(df_psl)

#Write to CSV File
df_psl.to_csv('PSL-Log.csv')
```
